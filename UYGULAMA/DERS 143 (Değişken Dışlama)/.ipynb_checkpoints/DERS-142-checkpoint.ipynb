```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.stats.diagnostic as smd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.DataFrame(pd.read_excel("ornek.xlsx"))
df.head()
```

```python
y = df["y"]
x = df[["x1","x2","x3","x4","x5","x6","x7","x8","x9","x10"]]

sabit = sm.add_constant(x)
model = sm.OLS(y,sabit).fit()
print(model.summary())
# Prob (F-statistic):           6.09e-06 anlamlı ama bağımsız değişkenlerde anlamsızlık varsa ilk şüphelenilmesi gereken
#şey ise Çoklu Doğrusal Bağlantı PROBLEMİ
```

```python
#korelasyon matrixi
kor = x.corr()
sns.heatmap(kor,annot=True,fmt=".2f") #ön gözlem gibi incelenebilir
plt.show()
```

```python
#VIF testi
vif = pd.DataFrame()
vif["Değişkenler"] = x.columns
vif["VIF"] = [variance_inflation_factor(sabit.values,i+1) for i in range(x.shape[1])]
print(vif)
#10 üzerindekiler sıkıntı da demek
```
