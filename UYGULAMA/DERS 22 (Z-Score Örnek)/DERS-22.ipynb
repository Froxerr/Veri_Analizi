```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0) # Altta veri adlı rastgele veriyi en son üretilene sabitliyoruz
veri = np.random.normal(35,2,40000) # Veri, ortalama değeri 35 ve standart sapması 2 olan
print(veri)
```

```python
sns.displot(veri,kde=True)
plt.title("Veri Dağılım Grafiği",fontsize=15,loc="right",c="red") #Loc, çizginin nerede yerleştirileceğini belirtir; 
#sağsa sağa, solsa sola
plt.xlabel("Veriler",fontsize=10,c="red") # Yatay bir yazı
plt.ylabel("Frekans",fontsize=10,c="red") # Dikey bir yazı
plt.axvline(x=np.mean(veri),linestyle="--",linewidth=2.5,label="Ortalama",c="red")#Dikey bir çizgi çekiyor
#x = bu çizginin nerede çizileceğini belirler
#linestyle = çizginin hangi tarzda çizileceğini tanımlar
#linewidth = çizginin kalınlığını ayarlar
#Label = çizginin adını belirler
plt.axvline(x=np.mean(veri) - np.std(veri),
            linestyle="--",
            linewidth=2.5,
            label="Ortalama -1 Standart Sapma",
            c="green")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veri) + np.std(veri),
            linestyle="--",
            linewidth=2.5,
            label="Ortalama +1 Standart Sapma",
            c="green")#Dikey bir çizgi çekiyor
plt.legend(loc=2,fontsize="xx-small") # Attığımız labeli gösteriyor
plt.show()
```

```python
veriz = (veri - np.mean(veri))/np.std(veri) # Z-Score formülünü yazıyoruz
sns.displot(veriz,kde=True) 
plt.title("Veri Dağılım Grafiği",fontsize=15,loc="right",c="red") #Loc, çizginin nerede yerleştirileceğini belirtir; 
#sağsa sağa, solsa sola
plt.xlabel("Veriler",fontsize=10,c="red") # Yatay bir yazı
plt.ylabel("Frekans",fontsize=10,c="red") # Dikey bir yazı
plt.axvline(x=np.mean(veriz),linestyle="--",linewidth=2.5,label="Ortalama",c="red")#Dikey bir çizgi çekiyor
plt.xlim(-3,3) # X'e yazıcağımız değerler -3 ile 3 sayılarını yaz diyoruz
#x = bu çizginin nerede çizileceğini belirler
#linestyle = çizginin hangi tarzda çizileceğini tanımlar
#linewidth = çizginin kalınlığını ayarlar
#Label = çizginin adını belirler
plt.axvline(x=np.mean(veriz) - np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            label="1 Standart Sapma",
            c="black")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veriz) + np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            c="black")#Dikey bir çizgi çekiyor


plt.axvline(x=np.mean(veriz) -2* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            label="2 Standart Sapma",
            c="blue")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veriz) +2* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            c="blue")#Dikey bir çizgi çekiyor

plt.axvline(x=np.mean(veriz) -3* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            label="3 Standart Sapma",
            c="green")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veriz) +3* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            c="green")#Dikey bir çizgi çekiyor

plt.legend(loc=2,fontsize="xx-small") # Attığımız labeli gösteriyor
plt.show()
```

```python
from scipy import stats 
veriz = stats.zscore(veri) #Z-Score buradan otomatik hesaplatıyoruz
sns.displot(veriz,kde=True) 
plt.title("Veri Dağılım Grafiği",fontsize=15,loc="right",c="red") #Loc, çizginin nerede yerleştirileceğini belirtir; 
#sağsa sağa, solsa sola
plt.xlabel("Veriler",fontsize=10,c="red") # Yatay bir yazı
plt.ylabel("Frekans",fontsize=10,c="red") # Dikey bir yazı
plt.axvline(x=np.mean(veriz),linestyle="--",linewidth=2.5,label="Ortalama",c="red")#Dikey bir çizgi çekiyor
plt.xlim(-3,3) # X'e yazıcağımız değerler -3 ile 3 sayılarını yaz diyoruz
#x = bu çizginin nerede çizileceğini belirler
#linestyle = çizginin hangi tarzda çizileceğini tanımlar
#linewidth = çizginin kalınlığını ayarlar
#Label = çizginin adını belirler
plt.axvline(x=np.mean(veriz) - np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            label="1 Standart Sapma",
            c="black")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veriz) + np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            c="black")#Dikey bir çizgi çekiyor


plt.axvline(x=np.mean(veriz) -2* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            label="2 Standart Sapma",
            c="blue")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veriz) +2* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            c="blue")#Dikey bir çizgi çekiyor

plt.axvline(x=np.mean(veriz) -3* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            label="3 Standart Sapma",
            c="green")#Dikey bir çizgi çekiyor
plt.axvline(x=np.mean(veriz) +3* np.std(veriz),
            linestyle="--",
            linewidth=2.5,
            c="green")#Dikey bir çizgi çekiyor

plt.legend(loc=2,fontsize="xx-small") # Attığımız labeli gösteriyor
plt.show()
```
