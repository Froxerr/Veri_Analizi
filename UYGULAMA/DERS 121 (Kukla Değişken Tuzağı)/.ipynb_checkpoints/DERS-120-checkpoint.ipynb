```python
import pandas as pd
import numpy as np
from sklearn import preprocessing as pr
```

```python
df = pd.DataFrame(pd.read_excel("ornek.xlsx"))
df2 = df.copy()
df2.head()
```

```python
#df2["KodCinsiyet"] = pr.LabelEncoder().fit_transform(df2["Cinsiyet"])
#df2["KodAraçRenk"] = pr.LabelEncoder().fit_transform(df2["AraçRenk"])
#df2["KodEğitim"] = pr.LabelEncoder().fit_transform(df2["Eğitim"])
for i in df2.columns:
    df2["Kod"+i] = pr.LabelEncoder().fit_transform(df2[""+i])
```

```python
df2.sort_values(by=["KodEğitim"]) #Rastgele kafasına göre sayı vermiş ama biz hiyeraşik bir düzen istiyorsak 
#bu başka bir dersin konusu
#Bu dersin konusu Cinsiyet ile KodCinsiyet arasında hiçbir bağın bulunmadığını anlatmaya çalışıyoruz
```

```python
onehotdonusum = pd.get_dummies(df2,columns=["Cinsiyet","AraçRenk"])
onehotdonusum #Ben bunları aralarında herhangi bir hiyeraşik düzen olmadan 0 ve 1 olarak 2 sütun olarak böldüm diyor 
#Araç içinse 0 ve 1 array olarak ben bunların arasında herhangi bir hiyeraşi olmadan böldüm diyor
```
